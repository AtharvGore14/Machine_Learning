```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv(r"D:\FY\Machine_Learning\01\homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel('Area(sq ft)')
plt.ylabel('Price(USD)')
plt.scatter(df.area, df.price, color='red', marker='+')
```

creating linear regression model

```python
reg = linear_model.LinearRegression()
reg.fit(df[["area"]],df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
#gives the slope of the line
```

```python
reg.intercept_
#gives the y intercept of the line
```

```python
#y = mx + b
135.78767123*3300 + 180616.43835616432

```

```python

d = pd.read_csv(r"D:\FY\Machine_Learning\01\area.csv")
d.head(3)
```

```python
p=reg.predict(d)
```

```python
d['prices'] = p
```

```python
%matplotlib inline
plt.xlabel('Area(sq ft)')
plt.ylabel('Price(USD)')
plt.scatter(df.area, df.price, color='red', marker='+')
plt.plot(df.area,reg.predict(df[["area"]]),color='blue')
```
